$R^{2}=0$: 모델이 종속변수의 변동성을 전혀 설명하지 못함!(평균값으로 예측하는 것과 같음)

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(perch_full, perch_weight, random_state=42)
```

위에서 본 것처럼 사이킷런은 특성을 만들거나 전처리하기 위한 다양한 클래스를 제공함.  
사이킷런에서는 이런 클래스를 변환기=transformer라고 부른다.(llm의 그 트랜스포머 아님)  

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures()
poly.fit([[2, 3]]) # PolynomialFeatures클래스의 fit() 메서드가 하는 일은 특성의 개수를 파악하는 것이 전부이다.
print(poly.transform([[2, 3]])) #따라서 훈련데이터와 특성 개수만 동일하다면 fit() 메서드는 아무값이나 전달해도 무관함
```

변환기는 입력 데이터로 변환하는 데 타깃 데이터가 필요하지 않음.  
[2, 3]이 6개의 특성을 가진 샘플 [1, 2, 3, 4, 6, 9]로 바뀌었음!

```python
poly = PolynomialFeatures(include_bias=False) # include_bias=False로 지정하지 않아도 사이킷런 모델은 자동으로 추가된 절편 항을 무시함
poly.fit([[2, 3]])
print(poly.transform([[2, 3]])) # 이 코드에서는 절편을 위한 1이 추가되지 않음
```

```python
poly = PolynomialFeatures(include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
print(train_poly.shape)
```

```python
poly.get_feature_names_out()
```

`PolynomialFeatures`가 특성을 9개 생성하는 이유는?

원본 데이터에는 3개의 특성(길이, 높이, 두께)이 있음. `PolynomialFeatures(include_bias=False)`는 기본적으로 `degree=2`(2차)까지의 모든 조합을 만듦.

**생성되는 9개 특성:**
1. `x0` (길이)
2. `x1` (높이)
3. `x2` (두께)
4. `x0²` (길이²)
5. `x0 × x1` (길이 × 높이)
6. `x0 × x2` (길이 × 두께)
7. `x1²` (높이²)
8. `x1 × x2` (높이 × 두께)
9. `x2²` (두께²)

**계산 공식:**
$n$개의 특성에서 degree=$d$일 때 생성되는 특성 개수는:
$$\binom{n+d}{d} - 1 = \binom{3+2}{2} - 1 = \frac{5!}{2! \times 3!} - 1 = 10 - 1 = 9$$

(`include_bias=False`이므로 -1)

만약 `degree=3`으로 설정하면 $x_0^3$, $x_0^2 \times x_1$ 같은 3차 항까지 포함되어 더 많은 특성이 생성된다!

```python
test_poly = poly.transform(test_input)
```

```python
print(lr.score(test_poly, test_target))
```

```python
poly = PolynomialFeatures(degree=5, include_bias=False)
poly.fit(train_input)
train_poly = poly.transform(train_input)
test_poly = poly.transform(test_input)
print(train_poly.shape)
```

```python
lr.fit(train_poly, train_target)
print(lr.score(train_poly, train_target))
```

```python
print(lr.score(test_poly, test_target))
```

너무 과대적합이라 오히려 test 데이터 값을 넣으면 이상한 값이 나온다!  
이런 과대적합을 줄이기 위해 규제. 즉, Regularization이 필요함!

### 규제  
규제는 머신러닝 모델이 훈련 세트를 너무 과도하게 학습시키지 못하도록 훼방을 놓는 것을 말한다.  
모델이 훈련  세트에 과적합되지 않게 하려고 하는 예방책임! 선형회귀 모델의 경우 특성에 곱해지는 계수의 크기를 작게 만드는 일

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_poly)
train_scaled = ss.transform(train_poly)
test_scaled = ss.transform(test_poly)
```

위 코드는 StandardScaler 클래스의 객체 ss를 초기화한 후 PolynomialFeatures 클래스로 만든  
train_poly를 사용해 이 객체를 훈련한다. 여기서도 강조되지만 꼭 훈련 세트로 학습한 변환기를 사용해 테스트 세트까지 변환해야 함

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
```

```python
print(ridge.score(test_scaled, test_target))
```

이때 Ridge와 lasso의 경우 규제의 양을 임의로 정할 수 있다.  
모델 객체를 만들 때 alpha의 값을 조절하는데 이 값이 크면 규제 강도가 세지고
이 값이 작으면 규제 강도가 작아진다! 이때 alpha의 값이 작으면 계수를 줄이는 역할이 줄어들고 선형 회귀 모델과 유사해져서  
과대적합이 될 가능성이 커진다! => 이때 alpha 값은 사람이 정해야 하는 Hyper-paramete이다

```python
import matplotlib.pyplot as plt
train_score = []
test_score = []
```

```python
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
    ridge = Ridge(alpha=alpha)
    ridge.fit(train_scaled, train_target)
    train_score.append(ridge.score(train_scaled, train_target))
    test_score.append(ridge.score(test_scaled, test_target))
```

```python
plt.plot(alpha_list, train_score)
plt.plot(alpha_list, test_score)
plt.xscale('log')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```

```python
# 위 그래프를 바탕으로 최선의 alpha값 선택
ridge = Ridge(alpha=0.1)
ridge.fit(train_scaled, train_target)
print(ridge.score(train_scaled, train_target))
print(ridge.score(test_scaled, test_target))
```

이번에는 lasso 회귀를 해보자!!  
ridge랑 매우 비슷하지만 Ridge클래스를 lasso클래스로 바꾸는 것이다!

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
```

```python
print(lasso.score(test_scaled, test_target))
```

```python
train_score = []
test_score = []
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100]
for alpha in alpha_list:
    lasso = Lasso(alpha=alpha, max_iter=10000)
    lasso.fit(train_scaled, train_target)
    train_score.append(lasso.score(train_scaled, train_target))
    test_score.append(lasso.score(test_scaled, test_target))
```

```python
plt.plot(alpha_list, train_score)
plt.plot(alpha_list, test_score)
plt.xscale('log')
plt.xlabel('alpha')
plt.ylabel('R^2')
plt.show()
```

```python
lasso = Lasso(alpha=10)
lasso.fit(train_scaled, train_target)
print(lasso.score(train_scaled, train_target))
print(lasso.score(test_scaled, test_target))
```

```python
print(np.sum(lasso.coef_ == 0)) # 이건 회귀 계수가 0인 특성의 개수를 출력. 즉, 라쏘에 의해 선택되지 않은 특성의 개수이다.
```

## 4-1 로지스틱 회귀

```python
import pandas as pd
fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head()
```

왜 여기서도 판다스의 데이터프레임을 사용하냐? 데이터프레임은 넘파이와도 잘 호환되고  
사이킷런과도 잘 호환된다.

```python
print(pd.unique(fish['Species']))
```

```python
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']]
```

```python
fish_input.head()
```

```python
fish_target = fish['Species']
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42) 
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_scaled, train_target)
print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```
